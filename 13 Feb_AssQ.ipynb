Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans: LookupError class is a superclass of KeyError and IndexError. This class is used to handle exceptions.

```python
# KeyError:
try:
    dict1 = {1:"Kaily", 2:"Kim"}
    print(dict1[3])
except KeyError as e:
    print("Key not found", e)
```

```python
# IndexError:
try:
    list1 = ["Amit", "Sumit", "Mohit"]
    print(list1[3])
except IndexError as e:
    print("Index not found", e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
# ImportError : This is a class used to handle exception when we try to import something which does not exist.
try:
    from crypt import pwd
except ImportError as e:
    print("Import Error", e)
```

```python
# ModuleNotFoundError : This is a class used to handle exception when we try to import a module which does not exist.
try:
    import kate
except ModuleNotFoundError as e:
    print("Import Error", e)
```

Q6. List down some best practices for exception handling in python. 

Ans: Some best practices are:
    
    1. Use specific errorhandling classes for handling exceptions.
    
    2. Pass meaningful messages while error handling.
    
    3. Always use logging to log information instead of using print statement.
    
    4. Avoid writing unnecessary exceptionhandlers.
    
    5. Prepare proper documentation of entire code.
